```python
#concat เลยไม่ได้เพราะค่าในไฟล์หลังๆที่ไม่ใช่ aov หายหมดเลย งั้นลอง drop column อื่นนอกจาก date ก่อนแล้วค่อย concat
#drop paid_type column
df_aov.drop(['paid_type'], axis = 1)
```

```python
#ก็ยังไม่ได้ แสดงว่าปัญหาอยู่ที่ตัว date มันซ้ำกัน
#ดังนั้นลองใหม่คือ pivot ทุก file แล้วค่อยมา concat
df_aov = df_aov.groupby('captured_date')['AOV'].agg(['sum'])
```
