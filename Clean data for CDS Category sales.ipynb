```python
df = df.rename({'Sum of Offline.1':'Offline_2022','Sum of Online.1':'Online_2022'}, axis=1)
```

```python
df
```

```python
def category(x):
    if x == 'Beauty' or x == 'Beauty Masstige':
        return 'Beauty & Personal care'
    elif x == 'Home' or x == 'Small App' or x == 'Tech':
        return 'Tech'
    elif x == 'Travel and Luggage':
        return 'Travel'
    elif x == 'Others' or x =='(blank)':
        return 'Others'
    elif x == 'Pets':
        return 'Pets'
    else:
        return 'Fashion'
```

```python
category('(blank)')
```

```python
df['category'] = df['Row Labels'].apply(lambda x : category(x))
```
