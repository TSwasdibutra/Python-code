## Install packages

```
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
#Autotokenizer = convert string into number (token) so that computer can understand and rate it
#torch = torch is need to use argmax function
import pandas as pd
import numpy as np 
```

## Download the pre-trained model

```
#Create tokenizer function
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
#Create a function that will give Sentiment Analysis Score
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```
#example token
ex1 = tokenizer.encode('Stop suggesting roll out plan or partial roll out plan, without consent from BU ', return_tensors='pt')
ex1
#tokenizer function คือแค่เปลี่ยน word to number, so that computer can understand
```

```
#model function คือจะบอก sentiment score
score = model(ex1)
score
```

```
#get the highest value from number from model function
#int คือเพื่อให้แสดงมาแต่ตัวเลข (integer)
int(torch.argmax(score.logits))+1
#ใส่บวก 1 เข้าไปเพื่อให้คะแนนเป็นระหว่าง 1 - 5 เพราะตอนนี้คะแนนเป็นระหว่าง 0-4 ซึ่งดูยาก
```

## Import Survey data

```
df = pd.read_csv('/work/Central Digital BU NPS Survey (Q1 2023)(1-33).csv')
df
```

```
#Keep only column that gonna use
df = df[['ID', 'BU', 'start_doing', 'stop_doing', 'continue_doing', 'other_comments']].copy()
df
```

```
#change colum name
df = df.rename({'BU':'bu'}, axis = 'columns')
df
```

```
df.dtypes
```

```
#need to do 2 steps
#1. create function that combine all the processes from tokenizer >> model >> show sentiment score
#2. for loop the above function across "stop" column
def sentiment_score(i):
    tokens = tokenizer.encode(i, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```
sentiment_score('Implementing initiatives without approval')
```

```
#for loop through 'start_doing' column in the dataframe using apply & LAMBDA
df['score_start'] = df['start_doing'].apply(lambda x: sentiment_score(x))
```

```
df
```

```
#for loop through 'stop_doing' column in the dataframe using apply & LAMBDA
df['score_stop'] = df['stop_doing'].apply(lambda x: sentiment_score(x))
```

```
df
```

```
#for loop through 'continue_doing' column in the dataframe using apply & LAMBDA
df['score_continue'] = df['continue_doing'].apply(lambda x: sentiment_score(x))
```

```
df
```

```
#for loop through 'continue_doing' column in the dataframe using apply & LAMBDA
df['score_other'] = df['other_comments'].apply(lambda x: sentiment_score(x))
df
```

```
df.to_csv('sentiment.csv', index=False)
```
