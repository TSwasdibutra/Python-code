```python
#Check for Multicollinearity between independent varibales (X) using correlation between X
    df2[['spend', 'newcustomer',
         'aov','session', 'order']].corr()
```

```python
#Show Correlation in heatmap
df2_corr = df2[['spend', 'newcustomer',
     'aov','session', 'order']].corr()
```

```python
sns.heatmap(df2_corr, annot = True)
```

```python
#the result show that we can use only 1 from spend, session, and order inour regression
#as correlation > 0.75 will lead to multicollinearity problem in our Multiple Regression Model
```

```python
#Try using all variables
lm2 = smf.ols(formula = 'rev ~ spend + newcustomer + aov + session + order', data = df2).fit()
```

```python
#remove new customer as p value > 0.05
reg2 = smf.ols(formula = 'rev ~ spend + aov + session + order', data = df2).fit()
```
