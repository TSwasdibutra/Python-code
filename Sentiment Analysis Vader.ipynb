```
#tqdm is just for showing status bar when running for loop
from tqdm.notebook import tqdm
```

```
#Loop the polarity score on my dataset
res = {}
for i, row in df.iterrows():
    text = row['What is something we should stop doing to improve your experience?']
    myid = row['ID']
    res[myid] = sia.polarity_scores(text)
```

```
res
```

```
#store the result in dataframe
pd.DataFrame(res)
```

```
#Transpose the data
vader = pd.DataFrame(res).T
vader
#Index of this dataframe is ID number
```

```
#Chnage Index column name to ID so vader can be merged with the original dataframe
vader = (
    vader.reset_index()
    .rename(columns={'index' : 'ID'})
)
vader
```

```
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```
#import pre-trained model
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```
#Run for Roberta model
tokenizer(ex)
```

```
encoded_ex = tokenizer(ex, return_tensors='pt')
```

```
output = model(**encoded_ex)
```

```
output
```

```
scores = output[0][0].detach().numpy()
```

```
scores = softmax(scores)
```
