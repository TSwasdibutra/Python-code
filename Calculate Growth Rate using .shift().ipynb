```python
df = pd.read_csv(r'D:\CDS Revenue.csv')
```

```python
df
```

```python
#remove row with 0 data 
df.query('revenue != 0')
```

```python
#reset index
df2 = (df.query('revenue != 0')
       .reset_index(drop = True)
      )
```

```python
df2
```

```python
#cal growth rate of revenue
#1. create lagged revenue
df2['past_rev'] = df2['revenue'].shift(1)
#shift(1) คือเลื่อนค่าไปทางขวาหนึ่งตำแหน่ง คือเลื่อนค่าไปข้างหน้า
#ถ้า shift(-1) คือเลื่อนค่าไปทางซ้ายหนึ่งตำแหน่ง คือเลื่อนกลับไปข้างหลัง
```

```python
df2
```

```python
#growth rate
df2['growth_rate'] = (df2['revenue'] - df2['past_rev'])*100 / df2['past_rev']
```

```python
df2
```

```python
#actually, you can calculate it without creating new column for lag variable
df3 = (df.query('revenue != 0')
       .reset_index(drop = True)
      )
```

```python
df3
```

```python
df3['growth_rate'] = (df3['revenue'] - df3['revenue'].shift(1)) * 100 / df3['revenue'].shift(1)
```

```python
df3
```

```python
#turn into % format
df3['growth_percentage'] = (df3['growth_rate']
                            .round(2)
                            .astype(str) + '%'
                           )
```
